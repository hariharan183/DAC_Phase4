```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
traffic=pd.read_csv("traffic.csv")
print(traffic)
```

```python
mean=np.mean(traffic)
print(mean)
```

```python
std=np.std(traffic)
print(std)
```

```python
from matplotlib import pyplot as plt
x = mean
y = std
plt.plot(mean,std)
plt.title("Line graph")
plt.ylabel('Y axis')
plt.xlabel('X axis')
plt.show()
```

```python
from matplotlib import pyplot as plt
x = mean
y = std
plt.bar(x,y,color = 'red')
plt.title('Result')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
